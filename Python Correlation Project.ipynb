```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
import numpy as np
```

```python
df= pd.read_csv(r'C:\Users\samue\Desktop\movies.csv')
```

```python
df.head()
```

```python
#check for missing data
for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print(col,pct_missing)
```

```python
#datatypes of columns 
df.dtypes
```

```python
#Replace the nan values with zeros then convert the colums to intergaers
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df.head()
```

```python

df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace = False, ascending =False)
```

```python
#scatter plot to show it
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('budget vs gross earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
#plot budget vs gross
sns.regplot(x='budget', y='gross',data=df, scatter_kws={'color': "red"},line_kws={'color': "purple"})
```

```python
df.corr(method='spearman', numeric_only=True)
```

```python
correlation_matrix = df.corr(method='kendall',numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation matrix for numeric feautures')
plt.xlabel('Movie Feautures')
plt.ylabel('Movie Feautures')
plt.show()
```

```python
df.head()
```

```python
df_numerized = df.copy()

for col in df_numerized.select_dtypes(include=['object', 'string']).columns:
    df_numerized[col] = df_numerized[col].astype('category').cat.codes
```

```python
print(df_numerized.dtypes)
print(df_numerized[['name']].head())
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Feautures')
plt.ylabel('Movie Feautures')
plt.xlabel('Movie Feautures')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs= correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```
